# Vizualizácie výsledkov k bakalárskej práci

```python
import pandas as pd
from matplotlib import pyplot as plt
from plotly import express as px
from plotly.subplots import make_subplots
import seaborn as sns
import plotly.graph_objects as go
```

## Vysoké školy


```python
df = pd.read_excel("analysis/results/vs.xlsx", index_col="VŠ")
df.head()
```

```python
figure = make_subplots(rows=4, cols=1, subplot_titles=("Flexibilita štúdia", "Predmety", "Výuka učiteľov", "Prístup učiteľov"))

figure.add_trace(go.Bar(x=df.index, y=df.flexibilita, name="flex"),
                row=1, col=1)

figure.add_trace(go.Bar(x=df.index, y=df.predmety, name="pred"),
                row=2, col=1)

figure.add_trace(go.Bar(x=df.index, y=df["učitelia_výuka"], name="uc_vyuk"),
                row=3, col=1)

figure.add_trace(go.Bar(x=df.index, y=df["podpora_študentov"], name="uc_prist"),
                row=4, col=1)

for i in range(1, 7):
    figure.update_yaxes(range=[-2.5, 1.5], row=i)

figure.update_layout(height=1400, width=1000, title_text="Vizualizácia spokojnosti po vysokých školách pre jednotlivé dimenzie")
figure.show()
```

## Fakulty

```python
fakulty = pd.read_excel("analysis/results/fakulty.xlsx")
fakulty.head()
```

```python
def plot_university(vs, vyska=1600, sirka=800):
    dimenzie = ("Flexibilita štúdia", "Predmety", "Výučba učiteľov", "Prístup učiteľov", "Hodnotenie", "Zručnosti", "Podpora študentov")
    dims = ("flexibilita", "predmety", "výučba_učiteľov", "prístup_učiteľov", "hodnotenie", "zručnosti", "podpora_študentov")
    
    figure = make_subplots(rows=7, cols=1, subplot_titles=dimenzie)
    
    df = fakulty[fakulty["VŠ"] == vs]
    
    for i in range(7):
        figure.add_trace(go.Bar(x=df["Fakulta"], y=df[dims[i]], name=dims[i]),
                row=i+1, col=1)
        yrange = [int(min(df[dims[i]])) - 1, int(max(df[dims[i]])) + 1]
        figure.update_yaxes(range=yrange, row=i+1)

    figure.update_layout(height=vyska, width=sirka, title_text=f"Vizualizácia spokojnosti pre fakulty {vs}")
    return figure
```

## Odbory

```python
odbory = pd.read_excel("analysis/results/odbory.xlsx")
odbory.head()
```

```python
def plot_faculty(df, fak):
    dimenzie = ("Flexibilita štúdia", "Predmety", "Výučba učiteľov", "Prístup učiteľov", "Hodnotenie", "Zručnosti", "Podpora študentov")
    dims = ("flexibilita", "predmety", "výučba_učiteľov", "prístup_učiteľov", "hodnotenie", "zručnosti", "podpora_študentov")
    
    figure = make_subplots(rows=7, cols=1, subplot_titles=dimenzie)
    
    fakulta = df[df["Fakulta"] == fak]
    nazov = fak+" "+list(df["VŠ"])[0]
    
    for i in range(7):
        figure.add_trace(go.Scatter(x=fakulta["Odbor"], y=fakulta[dims[i]] ,mode="markers", name=dims[i]),
                row=i+1, col=1)
        
    figure.update_layout(height=2400, width=800, title_text=f"Vizualizácia spokojnosti pre {nazov}")
    return figure

    
def plot_odbory(vs):
    vs = odbory[odbory["VŠ"] == vs]
    fakulty = set(vs["Fakulta"])
    
    res = {}
    for fak in fakulty:
        res[fak] = plot_faculty(vs, fak)
    return res
```

```python
uk = plot_odbory("UK")
```
